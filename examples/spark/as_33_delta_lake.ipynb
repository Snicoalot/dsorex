Delta Lake Tables are stored in the storage service as a folder. Content of the tables are stored in open-source Parquet formatted files.

```
%fs ls /FileStore/shared_uploads/1297906913.RANLY.NEI/paysum_data_delta/
```

The changes are saved in the object-store (or distributed file-system in this case) in the '_delta_log' folder.

The Delta log files serve as a single source of truth to the table.  The transaction log is checked to changes when the table is used.  More information about the transaction log can be found [here](https://www.databricks.com/blog/2019/08/21/diving-into-delta-lake-unpacking-the-transaction-log.html).

```
%fs ls /FileStore/shared_uploads/1297906913.RANLY.NEI/paysum_data_delta/_delta_log
```

We can see the content of a write-ahead log file.

```
%cat /dbfs/FileStore/shared_uploads/1297906913.RANLY.NEI/paysum_data_delta/_delta_log/00000000000000000008.json
```

The Delta Lake tables can be queried. Notice that this doesn't require any special database server, leveraging the concept of serverless. The query is processed on the client and not within the distributed-file-system service.

```
%sql

SELECT * FROM deltaTest.paysum;
```

Delta Lake Tables supports partitioning.  Partitions are stored in seperate files.

```
df_paysum = spark.read.csv(
  "dbfs:/FileStore/shared_uploads/1297906913.RANLY.NEI/paysum_data/data_1.csv",
  header=True,
  inferSchema=True
)

delta_file_folder_2 = "dbfs:/FileStore/shared_uploads/1297906913.RANLY.NEI/paysum_data_delta_v2/"
df_paysum.repartition(8).write.format("delta").mode("overwrite").save(delta_file_folder_2)
```

```
%sql

CREATE TABLE IF NOT EXISTS deltaTest.paysum_v2 USING delta LOCATION 'dbfs:/FileStore/shared_uploads/1297906913.RANLY.NEI/paysum_data_delta_v2/';
```

```
%fs ls /FileStore/shared_uploads/1297906913.RANLY.NEI/paysum_data_delta_v2/
```

Delta Lake Tables support specifing a column to drive the partitioning. Delta Lake Tables show the partitioning critiera in the name of child folders.  The child folders hold the content.

```
df_paysum_3 = spark.read.csv(
  "dbfs:/FileStore/shared_uploads/1297906913.RANLY.NEI/paysum_data/data_1.csv",
  header=True,
  inferSchema=True
)

delta_file_folder_3 = "dbfs:/FileStore/shared_uploads/1297906913.RANLY.NEI/paysum_data_delta_v3/"
df_paysum.write.partitionBy("type").format("delta").mode("overwrite").save(delta_file_folder_3)
```

```
%fs ls /FileStore/shared_uploads/1297906913.RANLY.NEI/paysum_data_delta_v3/
```

```
%fs ls /FileStore/shared_uploads/1297906913.RANLY.NEI/paysum_data_delta_v3/type=CASH_IN/
```
